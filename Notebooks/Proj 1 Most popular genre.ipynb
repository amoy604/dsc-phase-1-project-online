```python
import csv
from pandas import DataFrame
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
```

```python
df = pd.read_csv('imdb.title.basics.csv')
#import the genres data i'd like to use
```

```python
df = df.drop(columns=['tconst', 'start_year', 'runtime_minutes', 'original_title'])
df = df.dropna()
#dropping the columns that i dont want to work with.
```

```python
df = df.head(5000)
```

```python
df_genre = df.genres.str.split(",").tolist() #Getting rid of the commas and changing to a list.
df_genre
```

```python
genre_list = []
for genre in df_genre:
    try:
        genre_list.extend(genre)
    except:
        continue
```

```python
genre_counts = pd.Series(genre_list).value_counts()
```

```python
genre_counts = genre_counts.to_frame()
```

```python
fig = genre_counts.plot(kind='bar', figsize=(13, 6))
fig.set_title('How Many of Each Genre')
fig.set_xlabel('Genres')
fig.set_ylabel('Count')
```
